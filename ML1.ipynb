```python
import pandas as pd
import matplotlib.pyplot as plt #veri görselleştirme
```

```python
df = pd.read_csv('data/Wage.csv')
```

```python
df.describe() #sayısal olanları tanımladı
```

```python
df.head(10) #ilk on satırı göaterir.
```

```python
# Maaş-Yaş ilişkisi
df1=df[['age','wage']]

# girdi -> feature -> X
X= df1['age']

# çıktı -> output, label -> y
y= df1['wage']

#ortalama hesapla
y_mean=df1.groupby('age').mean() #mean=ortalama

#garfiği hazırla (boyutlandır), figzise=figür boyutu, ax=grafiğin kendisi, fig=subplotsun bize verdiği figür
fig, ax = plt.subplots(figsize=(8,8)) #Sırasıyla genişilik ve yükseklik

#datayı çiz
ax.scatter(X,y, color='purple') #scatter, dağılım grafiği yapar

#ortalamayı çiz eğri olarak göstermek için
#ax.plot(y_mean.index, y_mean) #Hata alıyorum

#garfiği formatla
plt.title('MAAŞ - YAŞ')

#x ve y ekseni için ayrı ayrı başlık yazıyor
plt.xlabel('YAŞ')
plt.ylabel('MAAŞ')
plt.show() #grafik üstünde çıkan yazıyı kaldırdı.
```

```python
# Maaş - Yıl karşılaştırması
df2= df[['year','wage']]

# girdi -> feature -> X
X= df2['year']

# çıktı -> output, label -> y
y= df2['wage']

#grafiği oluştur
fig,ax= plt.subplots(figsize=(8,8))

#datayı grafiğe koy
ax.scatter(X,y, color='black')

#ortalama hesapla
y_mean=df2.groupby('year').mean()

#ortalamayı grafiğe çiz !!!--Hata--!!!
#plt.plot(y_mean, y_mean.index, color='red')

plt.title('YIL - MAAŞ')

plt.xlabel('YIL')
plt.ylabel('MAAŞ')
#üstteki yazıyı kaldırdı
plt.show()
```

```python
#Maaş - Eğitim karşılaştırma
#boxplot
df3=df[['education','wage']]

X=df3['education']
y=df3['wage']

# eğitim üzerinden gruplandır
y_grouped = df3.groupby('education')
y_grouped.boxplot(subplots=False, figsize=(8,8), rot=90)

#başlıklar
plt.title("EĞİTİM - MAAŞ DURUMU")
plt.xlabel("EĞİTİM")
plt.ylabel("MAAŞ")
plt.show()
```
