```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='white')

from sklearn import decomposition
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
iris= datasets.load_iris()
X= iris.data
y=iris.target 
```

```python
#PCA altrmadan nce, dataset'i 3D olarak grelim:
fig = plt.figure(1, figsize=(6, 5))

plt.clf() #"clear figure" anlamna gelir. 

ax = fig.add_subplot(111, projection='3d')  # 3D alt grafii olutur
"""
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
rectangle= dikdrtgen,
elevation= ykseklik,Bu parametre, grafiin grsel olarak ne kadar eik olacan belirtir. 
azimuth= Bu parametre, grafiin grsel olarak ne kadar dndrleceini belirtir.
#plt.cla() "clear axes" anlamna gelir. 
bunlar almad
"""

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))

    
# Label'larn srasn ayn yapalm
y_clr = np.choose(y, [1, 2, 0]).astype(np.float_)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y_clr, 
           cmap=plt.cm.nipy_spectral)

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score

# Train, test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, 
                                                    stratify=y, #ylerin orann koru
                                                    random_state=42)

#datamz 3 boyutlu
X_train.shape
```

```python
# Decision tree, max_depth = 2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# modeli fit edelim
clf.fit(X_train, y_train)
```

```python
# tahmin oranlar
preds = clf.predict_proba(X_test) #probably=olaslk

print(preds)
```

```python
# accuracy - PCA olmadan

y_pred = clf.predict(X_test)

print('PCA Olmadan Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
```

Grdnz gibi 4 deiken ile model kurduumuzda accuracy deeri: %89 oldu.

imdi PCA ile deiken saysn azaltalm ve tekrar deneyelim:

```python
# sklearn'den PCA alalm
pca = decomposition.PCA(n_components=2) #n_components=datadaki dnmesini istediimiz deiken says
pca
```

Grdnz gibi imdi data artk 2 boyutlu ve 2D olduu iin de kolayca grafiini izdik.

```python
# Train, test split -> X_pca

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=.3, 
                                                    stratify=y, 
                                                    random_state=42)
```

```python
# Decision tree, max_depth = 2
clf_pca = DecisionTreeClassifier(max_depth=2, random_state=42)

# modeli fit edelim
clf_pca.fit(X_train, y_train)
```

```python
# accuracy - PCA olmadan

y_pred_pca = clf_pca.predict(X_test)

print('PCA ile Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred_pca)))
```

Her bir data noktas yani her bir rakam 8x8 pixellik bir alanda izilmitir.

Dolays ile her bir resmin boyutu 8*8 = 64'tr.

Her bir data noktas bu 64 adet pixelin arkaya arkaya yazlmas ile oluturulmutur. (flatten edilmi - dzletirilmi)

Her bir index'teki saylar da 0-255 arasnda bir deeri gsterir.

0 -> tam siyah
255 -> tam beyaz



**PCA'in faydas:**

Datamz 64 boyutlu yani boyut says fazla.

PCA ile boyut saysn 2 yapacaz.

Bylece data hem grsel olarak izilebilir ve ayrlabilir olacak hem de ilemlerimiz ok hzl olacak.

```python
# PCA nesnesi yaratalm
pca= decomposition.PCA(n_components=2)

# data boyutunu drelim, reduced=azaltlm, projecting=yanstma
X_reduced=pca.fit_transform(X)
"""
%d formatlama zellii, bir dize iinde yer tutucu olarak kullanlr ve bu yer tutucu, bir saysal deerle doldurulur.
Burada %d, bir tam sayy temsil eder.
"""
print("%d-boyutlu data 2 bpyutlu uzaya yanstld (projecting)."%X.shape[1])
```

```python
X_reduced.shape
```

```python
# imdi 2D'de izelim

plt.figure(figsize=(12,10))

#[:,0]  = bir matrisin veya bir veri dizisinin tm satrlarn ieren sadece ilk stununu alr.
plt.scatter(X_reduced[:,0],X_reduced[:,1], c=y,
           edgecolor="none", alpha=0.7, s=40, #edgecolor=kenar rengi, alpha=effaflk, s=noktalarn boyutu
           cmap=plt.cm.get_cmap("nipy_spectral",10)) #renk haritas
plt.colorbar()

plt.title("MNIST - PCA Projeksiyonu")
```

**PCA'in Gerek Hayatta Kullanm:**

Gerek hayat problemlerinde PCA yaplrken, genelde component says nden verilmez, bunun yerine toplam variance'in yzde kan aklayacak component'ler seilir.

rnein denir ki: Toplam variance'n %80'ini anlatan component'ler bizim iin yeterli.

te bu ekilde Principal Component'ler belirlenmi olur.

NNIST rneimizde, biz de diyelim ki, toplam variance'n %90'n anlatan componentleri bulmak istiyoruz. (explained_variance_ratio_)

Tahminimizi o component'ler ile yapacaz:

```python
# PCA ile toplam aklanan variance' izelim

pca= decomposition.PCA().fit(X)
plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_),color="k", lw=2)

plt.xlabel('Principal Component Says')
plt.ylabel('Toplam Aklanan Variance')

plt.xlim(0, 63)
plt.yticks(np.arange(0, 1.1, 0.1))

plt.axvline(21, c='b')
plt.axhline(0.9, c='r')

plt.show()
```

Grld gibi 21 adet deiken ile (Principal Component'ler) toplam variance'n %90' aklam oluyor.

Dolays ile artk 64 deiken ile deil, sadece 21 deiken ile bir model kurup hemen hemen ayn dorulukta tahminler yapabilir.

Yani tahmin kalitesinden pek birey kaybetmedik (max %10) ama performans'tan ok ciddi oranda kazanm elde ettik.
